```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.programmableweb.com/category/all/apis'
domain_url = 'https://www.programmableweb.com/'
```

```python
response = requests.get(url)
```

```python
data = response.text
```

```python
soup = BeautifulSoup(data, 'html.parser')
```

```python
tags = soup.find_all('a')
for tag in tags:
    print(tag.get('href'))
```

```python
# scraping names of apis 
titles = []
data = soup.find_all('td', {'class':'views-field-pw-version-title'})
for td in data:
    links = td.find_all('a')
    for a in links:
        titles.append(a.text)
titles
```

```python
# scraping api absolute urls
a_urls = []
data = soup.find_all('td', {'class':'views-field-pw-version-title'})
for td in data:
    links = td.find_all('a')
    for a in links:
        a_urls.append(domain_url[:-1]+a.get('href'))
a_urls
```

```python
# scraping category of each above
u_cat = []
data = soup.find_all('td', {'class':'views-field-field-article-primary-category'})
for td in data:
    links = td.find_all('a')
    for a in links:
        u_cat.append(a.text)
u_cat
```

```python
# Open the link of each api and extract its full  description
data = soup.find_all('td', {'class':'views-field-pw-version-title'})
abs_urls = []
for td in data:
    links = td.find_all('a')
    for a in links:
        abs_urls.append(domain_url[:-1]+a.get('href'))
u_descr = []
for uri in abs_urls:
    get_response = requests.get(uri)
    response_data = get_response.text
    api_soup = BeautifulSoup(response_data, 'html.parser')
    api_descr = api_soup.find('div',{'class':'api_description'}).text
    u_descr.append(api_descr[:20])
    
u_descr
```

```python
# combining all above 
set(zip(titles,a_urls,u_cat,u_descr))
```

```python
import requests
from bs4 import BeautifulSoup
# get page

url = 'https://www.programmableweb.com/category/all/apis'
domain_url = 'https://www.programmableweb.com'
while True:
    response = requests.get(url)
    data = response.text
    soup = BeautifulSoup(data,'html.parser')

    titles = []
    titles_data = soup.find_all('td', {'class':'views-field-pw-version-title'})
    for td in titles_data:
        links = td.find_all('a')
        for a in links:
            titles.append(a.text)
            
    a_urls = []
    a_urls_data = soup.find_all('td', {'class':'views-field-pw-version-title'})
    for td in a_urls_data:
        links = td.find_all('a')
        for a in links:
            a_urls.append(domain_url[:-1]+a.get('href'))
    
    
    u_cat = []
    cat_data = soup.find_all('td', {'class':'views-field-field-article-primary-category'})
    for td in cat_data:
        links = td.find_all('a')
        for a in links:
            u_cat.append(a.text)
            
            
    u_descr = []
    for uri in a_urls:
        get_response = requests.get(uri)
        response_data = get_response.text
        api_soup = BeautifulSoup(response_data, 'html.parser')
        api_descr = api_soup.find('div',{'class':'api_description'}).text
        u_descr.append(api_descr[:20]+'....')
            
            
            
    print(set(zip(titles,a_urls,u_cat,u_descr)))
    url_tag = soup.find('a',{'title': 'Go to next page'})
    
    if url_tag.get('href'):
        url= 'https://www.programmableweb.com' + url_tag.get('href')
        print("\nThis page url: "+url)
    else:
        break   
    
```
